```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create some random data
rng = np.random.default_rng()

# Generate an array of random numbers following a Gaussian distribution
example_array = rng.standard_normal(50)  

# create a figure object and ax
fig, ax = plt.subplots()

# plot the histogram of this array with
# the number of bins equal to the nearest integer value of the sqrt of the length of the array
nbins = int(np.sqrt(len(example_array)))
ax.hist(example_array, bins=nbins);

# add axes labels
ax.set_xlabel('value')
ax.set_ylabel('frequency')

# save the figure
fig.savefig('example_figure.png')
```
